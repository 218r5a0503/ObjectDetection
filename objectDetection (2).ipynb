```python
import cv2
cv2.version
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
from numpy.lib.polynomial import poly

from google.colab import files
data_to_load = files.upload()

imageName = list(data_to_load.keys())[0]
#Read the image using imread and display imshow functions
image = cv2.imread(imageName)
plt.imshow(image)
plt.show()

#Detect common objects using detect_common_objects function of cv2
box, label, count = cv.detect_common_objects(image)

#Draw the box around the detected objects and assign labels
output = draw_bbox(image, box, label, count)

#Dispay the image 
plt.imshow(output)
plt.show()

#Display the common objects detected 

object_counts = {}
for obj in label:
    if obj in object_counts:
        object_counts[obj] += 1
    else:
        object_counts[obj] = 1

# Display the count of each object detected in the image
for obj, count in object_counts.items():
    print(f"Number of {obj}s in this image: {count}")




```
